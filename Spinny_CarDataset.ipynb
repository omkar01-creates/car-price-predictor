```python
X = df[['insurance_validity', 'ownership', 'fuel_type', 'kms_driven', 'transmission']]
Y = df['price(in lakhs)']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
models = { 'LinReg': LinearRegression(),
           'KNN-3':KNeighborsRegressor(n_neighbors=3),
           'SVM-Poly':SVR(kernel='poly')}
for name, model in models.items():
    model.fit(X,Y)
    sc = model.score(X,Y)
    print(name,sc)
```

```python
d1 = {'Comprehensive': 0,'Zero Dep': 1,'Third Party': 3,'Third Party insurance': 2,'Not Available': 4}
d2 = {'First Owner': 1,'Second Owner': 2,'Third Owner': 3,'Fourth Owner': 4, 'Fifth Owner': 5}
d3 = {'Petrol': 0,'Diesel': 1,'CNG': 2}
d4 = {'Manual': 0,'Automatic': 1}
```

```python
insurance_validity = input('insurance_validity')
ownership = input('ownership')
fuel_type =input('fuel_type')
kms_driven = input('kms_driven')
transmission = input('transmission')
insurance_validity = d1[insurance_validity]
ownership = d2[ownership]
fuel_type = d3[fuel_type]
kms_driven = int(kms_driven)
transmission  = d4[transmission]
test = [[insurance_validity, ownership, fuel_type, kms_driven, transmission]]
yp = int(final_model.predict(test)[0])
print('Pridicted value is: {} Rs'.format(yp))
```
